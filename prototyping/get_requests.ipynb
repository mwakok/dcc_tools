```python
# Get parameters
import os
from requests import request
import json

# GitHub authentication
github_username = os.environ["GITHUB_USER"]
github_token = os.environ["GITHUB_TOKEN"]

# The repository to add this issue to
repo_owner = github_username
repo_name = "test"

headers = {"Authorization": f"token {github_token}",
            "Accept": "application/vnd.github.v3+json",
        } 
```

```python

def get_request(url, headers):
    """[summary]

    Parameters
    ----------
    url : [type]
        [description]
    headers : [type]
        [description]

    Returns
    -------
    [type]
        [description]
    """
    r = request("GET", url=url, headers=headers)
    if r.status_code == 200:
        print("Successfully retrieved data")       
        return json.loads(r.content)
    else:
        print("Could not get content", r.status_code)
        print("Response:", json.loads(r.content))
        return None
        

def get_issues(repo_owner, repo_name, headers):
    """[summary]

    Returns
    -------
    [type]
        [description]
    """
    url = f"https://api.github.com/repos/{repo_owner}/{repo_name}/issues"

    return get_request(url, headers)


def get_projects(repo_owner, repo_name, headers):
    """[summary]

    Returns
    -------
    [type]
        [description]
    """
    url = f"https://api.github.com/repos/{repo_owner}/{repo_name}/projects"

    return get_request(url, headers)
```

```python
issues = get_issues(repo_owner, repo_name, headers)
parsed_issues = {issue["number"]:issue["title"] for issue in issues}
parsed_issues
```

```python
projects = get_projects(repo_owner, repo_name, headers)
parsed_projects = {project["id"]:project["name"] for project in projects}
parsed_projects
```
